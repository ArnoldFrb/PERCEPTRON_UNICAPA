## EJECUTAR ENTRENAMIENTO

SE DEFINE LOS ARCHIVOS DE ENTRADAS Y SALIDAS DE LA NEURONA, TAMBIEN LA CONFIGURACION DE ESTA, COMO LA RATA DE APRENDIZAJE, ERROR MAXIMO, EL NUMERO DE ITERACIONES, LA FUNCION DE SALIDA QUE SE DESEA USAR Y EL NOMBRE DEL ENTRENAMIENTO

FUNCIONES DE SALIDA
- 1 => ESCALON
- 2 => LIENAL
- 3 => SIGMOIDE

ASEGURESE QUE LOS ARCHIVOS SEAN DEL TIPO TXT Y SE ENCUENTREN EN LA CARPETA DATA DEL PROYECTO

```python
from ConfigEntrenamiento import *

# NOMBRE DEL ARCHIVO DE ENTRADAS
ARCHIVO_ENTRADA = 'COMPUERTAS.TXT'

# NOMBRE DEL ARCHIVO DE SALIDAS
ARCHIVO_SALIDA = 'OR.TXT'

# ENTRENAMIENTO QUE DESEA REALIZAR
ENTRENAMIENTO = 'COMPUERTAS'

# NOMBRE DEL RESULTADO DEL ENTRENAMIENTO QUE DESEA REALIZAR
NOMBRE_DEL_ENTRENAMIENTO = 'COMPUERTAS_OR'

# RATA DE APRENDIZAJE DEL ENTRENAMIENTO
RATA_APRENDIZAJE = 1

# ERROR MAXIMO PERMITIDO PARA EL ENTRENAMIENTO
ERROR_MAXIMO = 0.001

# NUMERO MAXIMO DE ITERACIONES QUE REALIZARA EL ENTRENAMIENTO
NUMERO_ITERACIONES = 10000

# FUNCION DE SALIDA PARA EL ENTRENAMIENTO
FUNCION_SALIDA = 3

cfe = ConfigEntrenamiento(ARCHIVO_ENTRADA, ARCHIVO_SALIDA, ENTRENAMIENTO, NOMBRE_DEL_ENTRENAMIENTO)
cfe.ENTRENAR_NEURONA(RATA_APRENDIZAJE, ERROR_MAXIMO, NUMERO_ITERACIONES, FUNCION_SALIDA)
```
